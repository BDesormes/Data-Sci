```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,50) #defines x as func. that generates evenly spaced numbers over the specified interval: (0,10,50)
```

```python
noise = 10 * np.random.randn(50,) #defines noise as a func. that calculates the product of 10 times a random number within a range of (50,)
```

```python
y = 3 * x + 7 + noise #defines y
```

```python
plt.plot(x,y,'.'); #generates a scatter plot with the values of x, y in the prev functions
```

```python
m_guess = 1
b_guess = 1

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.');
```
